```python
prix=[485,491,488,495,493,502]
def calcul_rendement(prix_hier,prix_aujourdhui):
    r=(prix_aujourdhui-prix_hier)/prix_hier
    return r
rendement=[]
for i in range (1,len(prix)):
    rd=calcul_rendement(prix[i-1],prix[i])
    rendement.append(rd)
print(rendement)
import numpy as np
mu=np.mean(rendement)
sigma=np.std(rendement, ddof=1)
print(f"moyenne:{mu*100:.2f}%")
print(f"ecart type:{sigma*100:.2f}%")
mu_annuel=mu*252
sigma_annuel=sigma*np.sqrt(252)
sharpe=mu/sigma
sharpe_annuel=sharpe*np.sqrt(252)
print(mu_annuel*100,"%",sigma_annuel*100,"%",sharpe,sharpe_annuel)
```
