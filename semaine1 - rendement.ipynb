```python
import yfinance as yf
import numpy as np

# Télécharger et simplifier les données
apple = yf.download("AAPL", start="2024-01-01", end="2024-12-31")

# Aplatir les colonnes MultiIndex
apple.columns = apple.columns.get_level_values(0)

# Calculer les rendements
apple['rendement'] = apple['Close'].pct_change()
apple = apple.dropna()

# Statistiques
mu = apple['rendement'].mean()
sigma = apple['rendement'].std()
mu_annuel = mu * 252
sigma_annuel = sigma * np.sqrt(252)
sharpe = mu_annuel / sigma_annuel

print(f"Rendement annualisé : {mu_annuel*100:.1f}%")
print(f"Risque annualisé    : {sigma_annuel*100:.1f}%")
print(f"Sharpe annualisé    : {sharpe:.2f}")
print(f"Pire journée        : {apple['rendement'].min()*100:.2f}%")
print(f"Meilleure journée   : {apple['rendement'].max()*100:.2f}%")
```

```python
import matplotlib.pyplot as plt
plt.plot(apple.index,apple['Close'])
plt.title('evolution du cours Apple 2024')
plt.xlabel('date')
plt.ylabel('prix en USD')
plt.show
```
